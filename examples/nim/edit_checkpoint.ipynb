```python
new_cap = 75
```

```python
algs = checkpoint_manager.load_initial_algorithms()
```

```python
from decent_bench.schemes import Quantization
import torch
import numpy as np
import time

q = Quantization(4)
msg = np.random.randn(1000)
msg_torch = torch.from_numpy(msg)

start = time.time()
for _ in range(10000):
    __ = q.compress(msg)
print("Time taken for 10000 compressions:", time.time() - start)

start = time.time()
for _ in range(10000):
    __ = q.compress(msg_torch)
print("Time taken for 10000 compressions:", time.time() - start)
```
